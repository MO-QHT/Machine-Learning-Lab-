```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
pd.set_option("display.max_columns", None)

df = pd.read_csv("KFC_Past_Sales_Dataset.csv")
df.head(10)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df_missing = df.copy()
df_missing.loc[0:5, 'Sales'] = np.nan
df_missing.isna().sum()
```

```python
print("Original shape: ",df.shape)
print("After removing some values: ",df_missing.shape)
```

```python
df_missing.head(10)
```

```python
df_removed = df_missing.dropna()
df_removed.shape
```

```python
df_removed.isna().sum()
```

```python
df_missing.head(10)
```

```python
df_imputed_mean = df_missing.copy()
df_imputed_mean['Sales'].fillna(df_imputed_mean['Sales'].mean(), inplace=True)

df_imputed_mean.isna().sum()
```

```python
df_imputed_mean.head(10)
```

```python
df_imputed_median = df_missing.copy()
df_imputed_median['Sales'].fillna(df_imputed_median['Sales'].median(), inplace=True)

df_imputed_median.isna().sum()
```

```python
df_imputed_median.head(10)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Sales'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
Q1 = df['Customers'].quantile(0.25)
Q3 = df['Customers'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Customers'] < lower) | (df['Customers'] > upper)]
outliers.head(15)
```

```python
df_no_outliers = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
print("Original shape: ",df.shape)
print("After removing outliers: ",df_no_outliers.shape)
```

```python
lower_cap = df['Sales'].quantile(0.05)
upper_cap = df['Sales'].quantile(0.95)

df_capped = df.copy()
df_capped['Sales'] = df_capped['Sales'].clip(lower_cap, upper_cap)
```

```python
df[['Customers', 'Marketing_Spend']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df[['Sales', 'Marketing_Spend']].copy()

df_scaled[['Sales', 'Marketing_Spend']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df[['Sales', 'Marketing_Spend']].copy()

df_standardized[['Sales', 'Marketing_Spend']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales','Marketing_Spend']].corr(),annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df[['Sales', 'Marketing_Spend']].values

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
